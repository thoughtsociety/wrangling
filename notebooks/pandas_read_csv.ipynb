### Section 3 : Data I/O

#### Pandas Read and Write

One of the most common use-cases for pd.read_csv('file_name') is to import a 'flat', spreadsheet-style dataset into a Pandas dataframe.

We use this so we can then run:

* restructuring
* append
* imputation (adding data  to supplement missing data)
* aggregation
* row/column deletion
* joining
* merging
* scaling of numerical data

Of course, we frequently perform Exploratory Data Analysis (EDA) on the data once it is in a flat form.  It becomes much more straightforward to operate on when we have an idea of the mix of categorical and numerical data and their ranges.
